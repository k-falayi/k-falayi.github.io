##Nigeria's historic GDP growth rate

```python
df4 = pd.read_csv('gdp.csv')
df4.head()
```

```python
Nigeria = df4[df4['Country Name'].isin(['Nigeria'])]
Nigeria
```

```python
Nigeria = Nigeria.melt()
Nigeria.head()
```

```python
Nigeria = Nigeria.rename(columns={'variable':'year', 'value': 'GDP Percentage Change'})
```

### reduce dataframe to just 2012

```python
Nigeria = Nigeria.drop([0, 1, 2, 3])
```

```python
Nigeria['year'] = Nigeria['year'].astype(int)

```

```python
# Nigeria['GDP Percentage Change'] = Nigeria['GDP Percentage Change']
```

```python
Nigeria = Nigeria[Nigeria.year > 2001]
Nigeria
```

```python
Nigeria['GDP Growth Rate'] = Nigeria['GDP Percentage Change'] / 100
```

```python
Nigeria.head()
# Nigeria = pd.read_csv('NigeriaGDP.csv')
```

```python
chart = alt.Chart(Nigeria).mark_bar().configure(background='#353d4c').encode(
    alt.X("year:O"),
    alt.Y("GDP Growth Rate:Q", axis=alt.Axis(format='%')),
    color=alt.condition(
        alt.datum['GDP Percentage Change'] < 0,
        alt.value("red"), # The positive color
        alt.value("#6d98e8")  # The negative color
    ),
    tooltip=['GDP Percentage Change']
).properties(width=600)

chart = chart.configure_axisLeft(
  labelColor='white',
  titleColor='white',
).configure_axisBottom(
  labelColor='white',
  titleColor='white'
)



chart.save('NigGDP.html')
```
