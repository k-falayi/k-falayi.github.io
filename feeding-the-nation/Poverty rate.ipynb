```python
import pandas as pd
```

```python
df = pd.read_csv('all_csv/poverty_countries.csv')
df.head()
```

```python
df = df.rename(columns={'Entity': 'country'})
df = df.rename(columns={'$1.90 per day - share of population below poverty line': 'pct_belowPov'})
```

```python
df.head(2)
```

```python
df2 = pd.read_csv('country_continent.csv')
df2.head()
```

```python
new = df.merge(df2, on="country")
```

```python
new.head()
```

```python
df1 = new.query('continent == "Africa"')
df1.head()
```

```python
df1.groupby(['country','Year']).pct_belowPov.sum().to_frame().reset_index()
```
